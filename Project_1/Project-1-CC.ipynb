### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# On this challange, we treat the number (num) variable as a string.
# Expected result is: 906609 (913 x 993)

# Algorithm / Code Explanation
# 1. Conduct a nested iteration between the 3-digit numbers "x" and "y"
# 2. Determine if the product between the X and Y is palindromic
# 3. If it's a pali-number and higher than the "max" var.Assign "prod" to "max"
# 4. Repeat iterations until get's executed 999*999 times.
# 5. Present the results after loop

# Code has been verified against PEP8 (http://pep8online.com)


###############################
# Function Name: is_palindromic
# Function Purpose: To determine whether a number is palindromic or not
# Input: num (str)
# Output: True OR False (Bool)
##############################
def is_palindromic(num):
    num = str(num)  # convert to string for better

    if num == str(num)[::-1]:   # Determine if indexes (0, -1),
                                # (1, -2) and (2, -3) have the same value
        return True
    else:
        return False
# EOF is_palindromic()

max = 0  # Maximum Palindromic Value between the two two-digits number X and Y.
max_x = 0  # Maximum X to represent the result into a product form
max_y = 0  # Maximum Y to represent the result into a product form

# Cycle that executes product between two two-digits numbers (x, y)
# and verifies if is a palindromic
for x in range(100, 1000):
    for y in range(100, 1000):
        prod = x*y
        if is_palindromic(prod):
            if int(prod) > max:
                max = prod
                max_x = x
                max_y = y

# Output our information
print("Highest palindromic number is:", str(max), "(", max_x, "x", max_y, ")")

# EOF
```


### Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Expected result is: "The SUM of primes is: 277050"

# Algorithm / Code Explanation
# 1. Cycle all numbers between 2 and 1999
# 2. Determine if number in the loop iteration is prime or not
# 3. If prime (True) - add it to the "sum_primes" variables
# 4. Repeat iterations until 1999
# 5. Present the results

# Code has been verified against PEP8 (http://pep8online.com)


###############################
# Function Name: is_prime
# Function Purpose: To determine whether a number is prime or not
# Input: num (int)
# Output: True OR False (Bool)
##############################
def is_prime(num):
    if num > 1:
        n = num // 2
        for i in range(2, n + 1):
            if num % i == 0:
                return False
        return True
    else:
        return False

num = 2000  # set prime range
primes_sum = 0  # Sum of all primes (starts with zero)

for i in range(2, num):  # Cycle: from 2 to 1999
    if is_prime(i):
        primes_sum = primes_sum + i

# Print final output
print("The SUM of primes is: ", primes_sum)


# EOF
```

### Challenge 3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Expected result is: "The SUM is 233168"

# Algorithm / Code Explanation
# 1. Cycle all numbers between 2 and 1999
# 2. Determine if number in the loop iteration is prime or not
# 3. If prime (True) - add it to the "sum_primes" variables
# 4. Repeat iterations until 1999
# 5. Present the results

# Code has been verified against PEP8 (http://pep8online.com)


###############################
# Function Name: mult_three
# Function Purpose: To determine whether a number is multiple of 3 or not
# Input: num (int)
# Output: True OR False (Bool)
##############################
def mult_three(num):
    if (num % 3 == 0):
        return True
    else:
        return False


###############################
# Function Name: mult_five
# Function Purpose: To determine whether a number is multiple of 5 or not
# Input: num (int)
# Output: True OR False (Bool)
##############################
def mult_five(num):
    if (num % 5 == 0):
        return True
    else:
        return False

sum = 0  # sum of all multiples of three and fives. starts with zero
multiples = []

for i in range(1, 1000):  # loop all numbers between 1 and 999 (inclusive)
    if mult_three(i):
        multiples.append(i)
    if mult_five(i):
        multiples.append(i)

sum_dedup_list = list(set(multiples))     # Remove duplicates values (e.g: 975)

sum_dedup = 0  # SUM of values (counter)

for j in range(len(sum_dedup_list)):
    sum_dedup = sum_dedup + sum_dedup_list[j]

print("The SUM is", sum_dedup)
# EOF
```

### Challenge 4: String Compressor
Implement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# I am assuming that decoding of RLE is not part of this exercise.
# Encoding is CASE-SENSITIVE.

# Expected output result is:
# aabcccccaaa -> a2b1c5a3
# Kaboooooooomm -> K1a1b1o8m2
# aaaaaaaaaaaaaaaaaaaaaaaaaaa -> a27
# AaBbCCCaaa -> AaBbCCCaaa (not modified)
# abc -> abc (not modified)

# Algorithm / Code Explanation
# 1. RLE function gets the string to encode
# 2. RLE functions counts the number of consecutive chars (case sensitive)
# 3. At the end, RLE function validate if the encoded strings is not
#    same or longer than the original, otherwise it will assume the original
# 4. Present output to the user

# Code has been verified against PEP8 (http://pep8online.com)


###############################
# Function Name: rle_encode
# Function Purpose: Encodes a string into Run-Lenght-Encoding
# Input: input_text (string)
# Output: aux_str (string)
##############################
def rle_encode(input_text):

    input_text = str(input_text)  # input validation
    encoded_list = []
    encoded_str = ""
    count = 1
    char = ""
    aux_str = ""  # Because strings are immutable, we'll need this variable

    for i in input_text:

        if i != char:
            if char:
                entry = [char, count]
                encoded_list.append(entry)
            count = 1
            char = i
        else:
            count += 1
    else:
        entry = [char, count]
        encoded_list.append(entry)

    for item in encoded_list:
        encoded_str += str(item)

    # This code removes unwanted chars from the string
    for item in encoded_str:
        if (item != "[") and (item != "]") and (item != "'") and \
           (item != ",") and (item != " "):
            aux_str += item

    # This code prevents a longer encoded string than the original
    for item in encoded_str:
        if len(aux_str) >= len(input_text):
            aux_str = input_text + " (not modified)"

    # return encoded string
    return aux_str
# EOF rle_encode()


# Strings to Encode
inputs_list = ["aabcccccaaa",
               "Kaboooooooomm",
               "aaaaaaaaaaaaaaaaaaaaaaaaaaa",
               "AaBbCCCaaa",
               "abc"]

for input in inputs_list:
    print(input, "->", rle_encode(input))
# EOF
```

### *BONUS* Challenge: FizzBuzz
Write a program that prints all of the numbers from 1 to 100. For multiples of 3, instead of the number, print "Fizz;" for multiples of 5, print "Buzz." For numbers that are multiples of both 3 and 5, print "FizzBuzz." Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# On this exercise, the must use ELIF instead of IF to avoid
# on checking all conditions. We only need the first condition met
# A number is mathmetically multiple of another if their remainder is zero

# Code has been verified against PEP8 (http://pep8online.com)


###############################
# Function Name: check_mult
# Function Purpose: Check wether a number is multiple of 3 and/or 5
# Input: num (int)
# Output: output (str)
##############################
def check_mult(num):
    if (num % 3 == 0) and (num % 5 == 0):
        output = "FizBuzz"
    elif (num % 3 == 0):
        output = "Fizz"
    elif (num % 5 == 0):
        output = "Buzz"
    else:
        output = num

    return output
# EOF check_mult

for i in range(1, 101):
    print(check_mult(i))

# EOF
```
