#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

```python
# Is there a more pythonic way to do this?
movies['content_rating'] = movies['content_rating'].replace(["NOT RATED", "APPROVED", "PASSED", "GP"], "UNRATED")
movies['content_rating']
```

#### Convert the following content ratings to "NC-17": X, TV-MA.

```python
movies['content_rating'] = movies['content_rating'].replace(["X", "TV-MA"], "NC-17")
movies['content_rating']
```

#### Count the number of missing values in each column.

```python
# Sum of missing values per column - Will return 3 missing values in "content_rating"
movies.isnull().sum()
```

#### If there are missing values: examine them, then fill them in with "reasonable" values.

```python
# List records with any missing values
movies[movies.isnull().any(axis=1)]

# Fill missing values with "UNRATED" - In this case, there are only missing values in "content_rating"
movies.content_rating.fillna(value='UNRATED')
```

#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
# Is there any more pythonic way of doing this in a single line?
movies_above_2h = movies[movies.duration > 120]
movies_below_2h = movies[movies.duration < 120]

r1 = movies_above_2h.star_rating.mean()
r2 = movies_below_2h.star_rating.mean()

print("Above 2 hours movies are", round(r1-r2,3), "higher than below 2 hours movies")
```

#### Calculate the average star rating for each genre, but only include genres with at least 10 movies


#### Option 1: manually create a list of relevant genres, then filter using that list

```python
# Manually create a list of all genres with at least 10 movies
genres = ["Drama", "Comedy", "Action", "Crime", "Biography", "Adventure", "Animation", "Horror", "Mystery"]

# Filter using isin function
movies[movies.genre.isin(genres)].groupby('genre')['star_rating'].mean()

```

#### Option 2: automatically create a list of relevant genres by saving the value_counts and then filtering

```python
values = movies['genre'].value_counts()
genres_index = list(values[values > 10].index)
movies[movies['genre'].isin(genres_index)].groupby('genre')['star_rating'].mean()
```

#### Option 3: calculate the average star rating for all genres, then filter using a boolean Series

```python
# Exercise:
# Frist calculate the average "star_rating" for each genre and then
# filter "genre" with at least 10 movies using boolean series

my_values = movies.groupby('genre')['genre'].count() >= 10
movies.groupby('genre')['star_rating'].mean()[my_values]
```
































































#### Option 4: aggregate by count and mean, then filter using the count

```python
all_movies = movies.groupby('genre').star_rating.agg(['count', 'mean'])
all_movies[all_movies['count'] > 10].sort_values('count', ascending=False)
```

## Bonus

#### Figure out something "interesting" using the actors data!

```python
# On this exercise, we are getting the top rated actors. This is calculated by 
# caculating the average star_rating from each actor in all movies that he/she has participated
# to minimise the bias, we have only considered actors who have been in at least 5 movies.

import ast

# Get all the actors from the list.
actors = pd.DataFrame(columns=["avg_rating", "total_movies"])

for i, movie in movies.iterrows():
    # Convert CSV string into a Python list.
    movie["actors_list"] = ast.literal_eval(movie["actors_list"])
    
    # Get rating for movie.
    rating = movie["star_rating"]
    
    # Get all the actors (dictionaries are unique-keyed)
    for actor in movie["actors_list"]:
        
        if actor not in actors.index:
            # If the actor is not in the database yet, add them.
            actors.loc[actor] = (rating, 1)
        else:
            # If they are already in the set, calculate the new average.
            actors.loc[actor]["avg_rating"] *= actors.loc[actor]["total_movies"]
            actors.loc[actor]["avg_rating"] += rating
            actors.loc[actor]["total_movies"] += 1
            actors.loc[actor]["avg_rating"] /= actors.loc[actor]["total_movies"]
            
# Now, filter out actors with less than 5 movies.
actors = actors[actors.total_movies >= 5]

# Round column values
actors['avg_rating'] = actors['avg_rating'].round(2)
actors['total_movies'] = actors['total_movies'].astype(int)

# Sort by average rating.
actors = actors.sort_values(by='avg_rating', ascending=False)

# Print Top-10 actors from IMDB 
print(actors[0:10])
```
